Q2 : Making a Tic-Tak-Toe with random player turn and restart-game option

```python
#Everytime you run the code please restart the Kernal and clear output to avoid buffering situation 

import random

def print_board(brd):
    ind = 0 
    for i in range(5):
        if(i%2==0):

            for j in range(5):
                if(j%2!=0):
                    print("|",end =" ")
                else:
                    print(brd[ind],end =" ")
                    ind+=1
            print(" ")

        else:
            print("----------")
        
def move_possible(pos_avb):
    for i in range(9):
        if(pos_avb[i] == 0):
            return True
    return False
        
def move_x(brd,pos_avb,pos):
    
    for i in range(9):
        
        if(brd[i]==pos):
            
            brd[i] = 'x'
            pos_avb[i] = 1
            return True
        
    return False

def move_o(brd,pos_avb,pos):
    
    for i in range(9):
        
        if(brd[i]==pos):
            
            brd[i] = 'o'
            pos_avb[i] = 1
            return True
        
    return False

def win_x(brd):
    
    if(brd[0]=='x' and brd[1]=='x' and brd[2]=='x'):
        return True
    elif(brd[3]=='x' and brd[4]=='x' and brd[5]=='x'):
        return True
    elif(brd[6]=='x' and brd[7]=='x' and brd[8]=='x'):
        return True

    elif(brd[0]=='x' and brd[3]=='x' and brd[6]=='x'):
        return True    
    elif(brd[1]=='x' and brd[4]=='x' and brd[7]=='x'):
        return True
    elif(brd[2]=='x' and brd[5]=='x' and brd[8]=='x'):
        return True
    
    elif(brd[0]=='x' and brd[4]=='x' and brd[8]=='x'):
        return True
    elif(brd[2]=='x' and brd[4]=='x' and brd[6]=='x'):
        return True
   
    else:
        return False
    

def win_o(brd):
    
    if(brd[0]=='o' and brd[1]=='o' and brd[2]=='o'):
        return True
    elif(brd[3]=='o' and brd[4]=='o' and brd[5]=='o'):
        return True
    elif(brd[6]=='o' and brd[7]=='o' and brd[8]=='o'):
        return True

    elif(brd[0]=='o' and brd[3]=='o' and brd[6]=='o'):
        return True    
    elif(brd[1]=='o' and brd[4]=='o' and brd[7]=='o'):
        return True
    elif(brd[2]=='o' and brd[5]=='o' and brd[8]=='o'):
        return True
    
    elif(brd[0]=='o' and brd[4]=='o' and brd[8]=='o'):
        return True
    elif(brd[2]=='o' and brd[4]=='o' and brd[6]=='o'):
        return True
   
    else:
        return False

play = True
while (play):
    
        play = False
        board = [7,8,9,4,5,6,1,2,3]
        pos_avb = [0,0,0,0,0,0,0,0,0]

        x_turn = False
        o_turn = False

        turn = random.randint(0,1)
        if(turn==0):
            x_turn=True
        else:
            o_turn=True

        while (move_possible(pos_avb)==True and win_x(board)==False and win_o(board)==False):
            print(" ")
            print_board(board)
            print(" ")

            if(x_turn):
                while True:
                    pos_x = int(input("Where do you want to place 'x' - "))
                    print(" ")
                    if (pos_x > 9 or pos_x < 0 or move_x(board,pos_avb,pos_x) == False):
                        print("Sorry, please re-enter.")
                        continue
                    else:
                        o_turn=True
                        x_turn=False
                        break

            elif(o_turn):
                while True:
                    pos_o = int(input("Where do you want to place 'o' - "))
                    print(" ")
                    if (pos_o > 9 or pos_o < 0 or move_o(board,pos_avb,pos_o) == False):
                        print("Sorry, please re-enter.")
                        continue
                    else:
                        x_turn=True
                        o_turn=False
                        break

            else:
                print("Restart the Kernal")




        print_board(board)
        print(" ")

        if(win_x(board)==True):
            print("Congrats 'x' has won the game")
        elif(win_o(board)==True):
            print("Congrats 'o' has won the game")
        elif(move_possible(pos_avb)==False):
            print("The game has been drawn")
        else:
            print("Restart the Kernal ")
            
        print(" ")
        t = int(input("Press 1 if u want to play again ?"))
        if(t == 1):
            play = True

```
