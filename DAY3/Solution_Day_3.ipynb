Q1 : WAP for taking one number a input and then using if elif else concept, assign a grade to the input number from A,B,C,D,E, Fail in a formatted report card

```python
marks = int(input("Enter the assigned marks out of 100 - "))

print('|-------|')
if(marks >= 90):
    print("|   A   |")
elif (marks>=75 and marks < 90):
    print("|   B   |")
elif (marks>=60 and marks < 75):
    print("|   C   |")
elif (marks>=45 and marks < 60):
    print("|   D   |")
elif (marks>=35 and marks < 45):
    print("|   E   |")
else:
    print("Need imporvement")
    
print('|-------|')
```

Q2 : Guessing Team India Score in 20-20 Cricket - WAP gives you a random score of Team India in a cricket match, and a player has to guess the score given by computer in between 1 to 250. 

```python
from random import randint 

while True:
    scr = int(input("guess the score "))
    if(scr>=1 and scr<=250):
        break
    else:
        print("Reduce your expectation for 20-20 Cricket. Enter the score again")
        
computer_score = randint(1,250) 

dif = abs(scr - computer_score)

print("The score generated by computer - ",computer_score)
if ( dif <=10 and dif >=1):
    print("Close by, you are True Indian Fan!")
else:
    print("You are not a True Indian Fan! :P")
```

Q3 : Roy wants to change his profile picture on Facebook. Now Facebook has some restrictions over the dimension of pictures that we can upload.

Q4 : Ankit is in a maze. The command center sent him a string which decodes to come out from the maze.

```python
def split(word): 
    return [char for char in word] 

init_x = 0 
init_y = 0 
comand = input("Enter the Command for Ankit - ")
comand = comand.lower()
comman = split(comand)
for command in comman:
    if(command == "l"):
        init_x += -1
    elif (command == "r"):
        init_x += +1
    elif (command == "u"):
        init_y += 1
    elif( command == "d"):
        init_y -= 1
    else: 
        print("Enter valid input, L,R,U,D")
    
print("Ankit is here - (", init_x, init_y,")")
```
